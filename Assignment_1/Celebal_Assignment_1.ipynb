# 3. Handling Missing Values

### 3a. Identifying the missing values

```python
# df.isnull() returns a DataFrame of True/False values — 
# True where value is missing, False where value is present.
print(df.isnull())
# .sum() then counts all the True values per column (since True = 1, False = 0)
print(df.isnull().sum())
# to print columns which only have missing values we do the following
print('\nColumns with missing values')
print('-----------------------------')
print(df.isnull().sum()[df.isnull().sum()>0])
```

### 3b. Fill/Drop Missing Values
##### We only drop missing values when the number of missing values is less
##### In this case as the number of missing values are more we will not drop them as the meaning of the dataframe might change
##### We can drop the missing values using the function df.dropna()
##### Hence we will fill the missing values with appropriate values

#### Strategy to fill missing values

##### 1. discount -> as its datatype is float, it should be filled with numerical values, hence we can use median to do so as the column might have outliers due to which mean might give wrong values
##### 2. what_customers_said -> its datatype is str, hence we can fill it using 'No reviews'
##### 3. seller_name -> its datatype is also str, hence we can fill it using 'Unknown seller'
##### 4. videos -> its datatype is also str, hence we can fill it using 'No videos'
##### 5. seller_information -> its datatype is also str, hence we can fill it using 'No information available'
##### 6. variations -> its datatype is also str, hence we can fill it using 'No variations'

```python
df['discount'] = df['discount'].fillna(df['discount'].median())
df['what_customers_said'] = df['what_customers_said'].fillna('No reviews')
df['seller_name'] = df['seller_name'].fillna('Unknown seller')
df['videos'] = df['videos'].fillna('No videos')
df['seller_information'] = df['seller_information'].fillna('No information available')
df['variations'] = df['variations'].fillna('No variations')
# All the missing values have been filled
# To verify it:
print(df.isnull().sum())
print('\nMissing values after cleaning: ',df.isnull().sum().sum())
df # All missing values have been filled in the dataframe
```

# 4. Basic Operations - Filter & Select

### 4a. Rename

```python
# Renaming the column what_customers_said to customer_reviews
df.rename(columns={"what_customers_said":"customer_reviews"},inplace=True)
df
```

### 4b. Selecting specific columns
##### As the DataFrame has 24 columns, out of those we can select a few meaningful columns and work with them
##### We can select columns like : product_id,title,product_description,rating,ratings_count,initial_price,discount,final_price,currency,delivery_options,product_details,seller_name,category

```python
new_df = df[['product_id', 'title', 'product_description', 'rating', 'ratings_count', 
         'initial_price', 'discount', 'final_price', 'currency', 'delivery_options', 
         'product_details', 'seller_name', 'category']]
print(f"New shape: {new_df.shape}")
print(f"Selected columns: {list(new_df.columns)}")
new_df.head()
```

### 4c. Filtering rows

```python
# Filtering products with rating >= 4.0
high_rated = new_df[new_df['rating'] >= 4.0]
print('High-rated products (≥ 4.0): ',len(high_rated))

# Filtering products with discount > 50%
high_discount = new_df[new_df['discount'] > 50]
print('Products with >50% discount: ',len(high_discount))

# Filtering products with price under ₹1000
budget = new_df[new_df['initial_price'] < 1000]
print('Budget products (< ₹1000): ',len(budget))

# Filtering according to the best deals using both high rated and high discount
best_deals = new_df[(new_df['rating'] >= 4.0) & (new_df['discount'] > 50)]
print('Best deals (rated ≥4.0 + discount >50%): ',len(best_deals))
best_deals.head()
```

### 4d. Dropping columns
##### Dropping the column final_price as the values are the same as initial_price,
##### meaning no discount has been applied, making it redundant

```python
new_df = new_df.drop(columns=['final_price'])
print('New shape: ',new_df.shape)
new_df
```

# 5. Remove Duplicates

### 5a. Checking if duplicates exist

```python
print('Number of duplicate rows: ', new_df.duplicated().sum())
print('Total rows before removing duplicates: ', len(new_df))
```

### 5b. Removing duplicates
##### In this case there are no duplicates but if they exist then we use the following code to remove them

```python
new_df = new_df.drop_duplicates()
print('Total rows after removing duplicates: ', len(new_df))
```

# 6. Derived Column Creation
##### Creating a derived column :    total_amount_to_be_paid=initial_price-(initial_price*(discount/100))

```python
new_df['total_amount_to_be_paid'] = new_df['initial_price'] - (new_df['initial_price'] * (new_df['discount'] / 100))
# Verify if the derived column is created or not
print(new_df[['initial_price', 'discount', 'total_amount_to_be_paid']].head())
```
